## Task 1: Image Processing in Python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('image1.jpg')
img2 = cv2.imread('image2.jpg')

images = [img1, img2]
titles = ['Image 1', 'Image 2']
```

```python
for img, title in zip(images, titles):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    plt.figure(figsize=(5,5))
    plt.imshow(gray, cmap='gray')
    plt.title(f'{title} - Grayscale')
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(5,3))
    plt.hist(gray.ravel(), bins=256)
    plt.title(f'{title} - Histogram')
    plt.show()

    resized = cv2.resize(img, (300,300))
    plt.figure(figsize=(5,5))
    plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
    plt.title(f'{title} - Resized')
    plt.axis('off')
    plt.show()

    edges = cv2.Canny(gray,100,200)
    plt.figure(figsize=(5,5))
    plt.imshow(edges, cmap='gray')
    plt.title(f'{title} - Edge Detection')
    plt.axis('off')
    plt.show()

    noise = np.random.normal(0,25,gray.shape)
    noisy = np.clip(gray + noise,0,255).astype(np.uint8)

    plt.figure(figsize=(5,5))
    plt.imshow(noisy,cmap='gray')
    plt.title(f'{title} - Noisy')
    plt.axis('off')
    plt.show()

    denoised = cv2.GaussianBlur(noisy,(5,5),0)

    plt.figure(figsize=(5,5))
    plt.imshow(denoised,cmap='gray')
    plt.title(f'{title} - Gaussian Filter')
    plt.axis('off')
    plt.show()
```

# Task 2: Medical Imaging Dataset Exploration

## Dataset 1: NIH Chest X-ray Dataset

- Type: Chest X-ray
- Number of Images: ~112,000
- Labels: 14 disease categories
- Dataset Imbalance: Rare diseases such as Hernia have fewer samples.
- Challenges: Label noise, class imbalance, visually similar diseases.

### Summary
One of the largest public chest X-ray datasets used for disease classification research.

---

## Dataset 2: ISIC Skin Cancer Dataset

- Type: Dermoscopic skin images
- Number of Images: 25,000+
- Labels: Melanoma, Nevus, Benign lesions, etc.
- Dataset Imbalance: Benign cases dominate the dataset.
- Challenges: Hair artifacts, lighting variation, image quality differences.

### Summary
A benchmark dataset for skin lesion classification and early skin cancer detection research.
